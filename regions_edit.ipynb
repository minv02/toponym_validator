```python
import pandas as pd
```

```python
regions = pd.read_csv("regions.csv")
regions.info()
```

# Правлю название мурманска

```python
old_murmansk_name = "Мурманск (город-герой)"
new_murmansk_name = "Мурманск"
if not regions[regions.town == old_murmansk_name].empty:
    regions.loc[regions.town == old_murmansk_name, "town"] = new_murmansk_name

regions[regions.town == new_murmansk_name]
```

# Добавляю города федерального значения

```python
federal_cities = [
    "Москва",
    "Санкт-Петербург",
    "Севастополь"
]

rows = []
for federal_city in federal_cities:
    row = [federal_city, federal_city, None, None, None]
    rows.append(row)

add_regions = pd.DataFrame(data=rows, columns=regions.columns)
regions = pd.concat([regions, add_regions])
regions.tail(10)
```

# Записываю все в новый файл

```python
regions.to_csv("regions.csv", index=False)
```
